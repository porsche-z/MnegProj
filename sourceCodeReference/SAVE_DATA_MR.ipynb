```python
import numpy as np
train_data = []
train_label = []
test_data = []
test_label = []
for i in range(45):
    
        
    train_data.extend(data_dict['IQdata'].T[i*400: (i+1)*400][:])
    test_data.extend(data_dict['IQdata'].T[i*400: (i+1)*400][:])
    if i != 22:
        train_label.extend(np.zeros((400,)))
        test_label.extend(np.zeros((400,)))
    else:
        train_label.extend(np.ones((400,)))
        test_label.extend(np.ones((400,)))
```

```python
train_label = np.asarray(train_label)
test_label = np.asarray(test_label)
train_data = np.asarray(train_data)
test_data = np.asarray(test_data)
print('shape of training/ testing set:', np.shape(train_data), np.shape(train_label), np.shape(test_data), np.shape(test_label))
```

```python
np.save('train_data',train_data)
np.save('train_label',train_label)
np.save('test_data',test_data)
np.save('test_label',test_label)
```
